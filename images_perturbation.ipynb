```python
from torchvision import transforms
from helpers import SemiSupervisedDataset
from helpers import display_images_in_rows
```

# Set paths

```python
DATA_PATH = './15SceneData/'
```

# Load perturbed data

```python
train_dataset_rotation = SemiSupervisedDataset(
    DATA_PATH + 'train',
    image_transformation='rotation',
    transform=transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        # transforms.Normalize((0.5,), (0.5,))
    ])
)
train_dataset_box = SemiSupervisedDataset(
    DATA_PATH + 'train',
    image_transformation='black-white',
    transform=transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        # transforms.Normalize((0.5,), (0.5,))
    ])
)
```

# Visualise perturbations

```python
num_images_to_display = 12
images_to_display = [transforms.functional.to_pil_image(train_dataset_rotation[i][0]) for i in range(num_images_to_display)]

# Display the images
display_images_in_rows(images_to_display, max_images_per_row=4, width=500, margin=5)
```

```python
num_images_to_display = 12
images_to_display = [transforms.functional.to_pil_image(train_dataset_box[i][0]) for i in range(num_images_to_display)]

# Display the images
display_images_in_rows(images_to_display, max_images_per_row=4, width=500, margin=5)
```
